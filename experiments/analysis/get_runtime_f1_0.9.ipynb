```python
import json
import numpy as np
```

```python
# PATSQL
# open json file
queries = ["(BottomQuadrant(o0), Near_1.0(o0, o1))",
            "Far_3.0(o0, o1); Near_1.0(o0, o1); Far_3.0(o0, o1)",
            "Far_3.0(o0, o1); (Behind(o0, o1), Near_1.0(o0, o1))",
            "Far_3.0(o0, o1); (Behind(o0, o1), LeftQuadrant(o0), Near_1.0(o0, o1))",
            "(FrontOf(o0, o1), TopQuadrant(o0))",
            "Near_1.0(o0, o1); Far_3.0(o0, o1)",
            "(Behind(o0, o1), LeftQuadrant(o0), Near_1.0(o0, o1))",
            "(BottomQuadrant(o0), Far_3.0(o0, o1)); Near_1.0(o0, o1)",
            "(Far_3.0(o0, o1), LeftQuadrant(o0)); (LeftQuadrant(o0), Near_1.0(o0, o1))"]
for query in queries:
    print(query)
    with open('/gscratch/balazinska/enhaoz/complex_event_video/src/quivr/outputs/without_duration-sampling_rate_4/stats/PATSQL/{}.json'.format(query)) as f:
        data = json.load(f)
    scores = data['score']
    runtimes = data['runtime']
    n_examples = list(range(12, 21)) + list(range(25, 51, 5))
    result_time = []
    result_examples = []
    for score_per_run, runtime_per_run in zip(scores, runtimes):
        for n_examples_per_run, score, runtime in zip(n_examples, score_per_run, runtime_per_run):
            if score > 0.9:
                result_time.append(runtime)
                result_examples.append(n_examples_per_run)
                break
    print(result_time)
    print("median:", np.median(result_time))
    print(result_examples)
    print("median:", np.median(result_examples))
```
