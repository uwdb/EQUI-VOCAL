```python
import cv2
import json
from matplotlib import pyplot as plt
```

```python
filename = "traffic-010"
frame_id = 1
video = cv2.VideoCapture("/home/ubuntu/complex_event_video/visualroad/my-dataset/{}.mp4".format(filename))

video.set(cv2.CAP_PROP_POS_FRAMES, frame_id)

ret, frame = video.read()

with open("/home/ubuntu/complex_event_video/visualroad/my-dataset/{}.json".format(filename), 'r') as f:
    bbox_json = json.loads(f.read())

for row in bbox_json["frame_{}.jpg".format(frame_id)]:
    x1, y1, x2, y2, class_name, _ = row 
    cv2.rectangle(frame, (int(x1), int(y1)), (int(x2), int(y2)), (0, 255, 0), 3)

frame = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB) # Converting BGR to RGB

plt.imshow(frame)
```
